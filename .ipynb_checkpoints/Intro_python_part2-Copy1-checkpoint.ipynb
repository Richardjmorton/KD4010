## Line plots
There are a number of ways you can generate plots with matplotlib, and how you do this depends on the context. First, we will deal with creating plots from the command line.


We can now create a simple plot that will appear as a pop-up window in a terminal or in a separate tab within Pycharm.

First, we will create a Figure object which opens up an empty window.

```python
x=np.arange(0,10,0.1) # array from 0 to 10 in steps of 0.1
y=np.sin(x)
y2=np.cos(x)

fig=plt.figure()  # Create figure object
```

Now we can plot to the window. We will do a simple line plot first.

```python
plt.plot(x,y,'-')
plt.plot(x,y2,'--')
```

You can see that if we keep making calls to plt.plot(), then the results will be over-plotted on the open figure window. You should also notice that I selected the style of the line with the dashes between the quotation marks.

We can save our figure with the following:

```python
fig.savefig('myfigure.png')
```

If no filepath is specified, e.g., '*my\_folder/pictures/myfigure.png*' then the above command just saves to the directory in which you launched Python (the current working directory).

You can also create multiple subplots in a figure

```python
plt.style.use('seaborn-whitegrid') # Change plot theme

fig=plt.figure()  # Create figure object

ax1=fig.add_subplot(1,2,1) # This creates 1 by 2 subplots in the window, and selects the first plot
ax2=fig.add_subplot(1,2,2) # This selects the second plot
```

We can now refer to the ax1 and ax2 objects to plot to the different subplots

```python
fig=plt.figure()  # Create figure object

ax1=fig.add_subplot(1,2,1) # This creates 1 by 2 subplots in the window, and selects the first plot
ax2=fig.add_subplot(1,2,2) # This selects the second plot

ax1.plot(x,y,'ro') # 'ro' gives red dots
ax2.plot(x,y2,linestyle='dashdot',color='g') # alternative method for specifiying color and line-style
```

You may want to change the axis ranges, or add titles

```python
fig=plt.figure()  # Create figure object

ax1=fig.add_subplot(1,2,1) # This creates 1 by 2 subplots in the window, and selects the first plot
ax2=fig.add_subplot(1,2,2) # This selects the second plot

ax1.plot(x,y,'ro') # 'ro' gives red dots
ax2.plot(x,y2,linestyle='dashdot',color='g') # alternative method for specifiying color and line-style

ax1.set_xlim(0,20)
ax1.set_ylim(-2,2)
ax1.set_xlabel('X')
ax1.set_ylabel('Ys')
ax1.set_title('Curves')
```

It is also possible to add a legend to the chart, but labels need to specified when calling plot.

```python
fig=plt.figure()  # Create figure object
plt.plot(x,y,label='sin')
plt.plot(x,y2,label='cos')
plt.legend()
```

## Scatter plots

While you can create scatter plots with the *plot* function, a more powerful option is *plt.scatter* .

This function enables you to control various aspects of each individual point, such as colour and size, to enable multi-dimensional information to be shown 

```python
# This example is taken from J. VanderPlas 'Python Data Science Handbook'
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

Note that the color argument in *scatter* is automatically linked to the colourbar. Also, the size of the data point is given in pixels. Finally, to make all the overlapping data points visible we have changed the transparency of the plot using the *alpha* argument.
