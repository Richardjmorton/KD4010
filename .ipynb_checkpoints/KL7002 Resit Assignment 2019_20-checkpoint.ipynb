
**(e)** While the plot provides a visual comparison between the two distributions, we want to quantify any difference between the two distributions. We are going to use a statistical test called the Kolmogorov-Smirnov (KS) two-sample test.

Provide a description of what the KS two-sample test does. State any assumptions behind the test, including its null and alternative hypothesis. Implement the test in Python (it is available in the SciPy stats package) to compare the two distribution of sunspot number. What do the results suggest about the two distributions?

**15 marks**

The KS two-sample test is a non-parametric test that aims to compare whether two samples are from the same population distribution.

The test looks for the maximum deviation between the two normalised cumulative distribution functions, $S_1(x)$ and $S_2(x)$, of length $m$ and $n$ respectively. The deviation is  
$$
D = max \mid S_2(x)- S_1(x)\mid \,.
$$

This is then compared to a critical value for the given sizes, $m$ and $n$, found from standard statistical tables.

(5 marks)

There are no assumptions on the distributions under examination.

The null hypothesis, $H_0$, is that the two samples are from the same population.

The alternative hypothesis $H_1$, is that the two samples differ from each other.

The test can either be one or two tailed.

A typical cut-off value is at the 5% level for a one tailed test (2.5% for a two-tailed test).

(5 marks)

````python
#Two tailed test
d, p =stats.ks_2samp(dnsn, dssn)

print('Maximum deviation {0:2.3f}, p-value {1}'.format(d, p))
````

The p-value returned is much smaller than 2.5%, which means that there is evidence to reject the null hypothesis. This means that the distributions of the northern and southern sunspot counts do not come from the same population distribution.

**(b)** Re-scale the dynamic range of the image by using a gamma transform/correction so that you can see the features better.

Produce a plot similar to that in Task 1, which shows the new image of the data and the new histogram of the pixel values side by side.

Discuss the gamma transform, why SDO/AIA image has poor contrast, and quantify how the technique has changed the dynamic range of the SDO/AIA images and improved the contrast.

**10 Marks**

````python
data = np.clip(data,1, data.max())

plt.figure(figsize=(10,10))
plt.imshow(data**0.2, extent=ax_range, origin='lower')
plt.xlabel('Solar X (arcsec)')
plt.ylabel('Solar Y (arcsec)')

plt.title(title)
````

5 marks for figure


The gamma transform is a nonlinear mapping of intensity values that can be used to adjust the contrast in images. The transform had the following form
$$
I_{out}=I_{in}^\gamma
$$
where $\gamma$ is the factor that is changed to alter the contrast.

SDO had poor contrast as there are a few large count values that are greater than 5000, while the majority are less than this. These count values have to be mapped onto the standard 8bit image range, which is from 0 to 255. Hence, most of the values are mapped onto only a small set of values in this range. We say that the AIA image has a large dynamic range.

The gamma transform shrinks the dynamic range, meaning that the count values are spread over a narrower range. The following histogram shows the distribution of intensity values following the non-linear transformation. 

````python
plt.hist( (data.flatten())**0.3, bins='scott', histtype='step')
plt.xlabel('Counts (DN)')
plt.ylabel('frequency')
plt.yscale('log')
````
