d) (i) 5 Marks  (ii) 10 Marks

(i)
Equation to use

$$
\sqrt[4]{ \frac{4\pi\sigma R^2_\odot}{L_\odot} }T=\left(\frac{L}{R^2}\right)^{0.25}
$$

where $L$ and $R$ are now in terms of solar luminosity and mass respectively.

```python
#Select stars
df_stars=df.dropna(subset=['st_spstr','st_teff','st_rad','st_lum'])

g_stars=df_stars[df_stars['st_spstr'].str.contains('G')]
k_stars=df_stars[df_stars['st_spstr'].str.contains('K')]

g_and_k_stars=pd.concat([g_stars,k_stars])
```

(ii)

```python
# Fit straight line to data
def lin_func(x,a,b):
    return x*b+a

x=g_and_k_stars.st_teff.values
lr2=10**(g_and_k_stars.st_lum.values)/g_and_k_stars.st_rad.values**2
y=lr2**0.25
popt, pcov = curve_fit(lin_func, x, y)
```

```python
print('Fitting results')
print('Constant {0:1.3g}, Gradient {1:1.3g}'.format(popt[0],popt[1]) )
```

```python
# plot scatter plot and over plot straight line fit

plt.scatter(x,y,alpha=0.3,facecolors='none',edgecolors='b')
plt.xlabel('$T_{eff}$ (K)')
plt.ylabel('$(\,{L}\, /\,{R^2})^{0.25}$ $(L_\odot /R^2_\odot)$')
plt.plot(x,x*popt[1]+popt[0],'-r',)
```

```python
# Calculate value of Stefan-Boltzmann constant
stef_const=(popt[1]/u.Kelvin)**4/4/np.pi*const.L_sun/const.R_sun**2
print('Stefan-Boltzmann constant is {0:1.3g}'.format(stef_const))
```

```python

home = os.path.expanduser("~")

filepath=os.path.join(home,'analysis/SDO/hmi_level15/HMI20131217_130024_mag.fits')
with fits.open(filepath) as hdul:
    data=hdul[0].data
    header=hdul[0].header
```

```python
r_sun_pixels=header['RSUN_OBS']/header['CDELT1']
ny,nx=data.shape
```

```python
x=np.arange(0,nx)-nx/2+0.5
x=x[:,np.newaxis]*np.ones(nx)
r_polar=(x**2+x.T**2)**0.5
data[(r_polar>(r_sun_pixels))]=0

axis_labels=((np.arange(0,nx)-header['CRPIX1']-1))*header['CDELT1']
mn_ax=np.min(axis_labels)
mx_ax=np.max(axis_labels)
plt.figure(figsize=(10,10))
plt.imshow(np.clip(data,-100,100),cmap='gray',extent=[mn_ax, mx_ax, mn_ax, mx_ax],origin='lower')
plt.xlabel('Solar X (arcsec)')
plt.ylabel('Solar Y (arcsec)')
title_string=[header['TELESCOP'],header['INSTRUME'],header['CONTENT'],header['DATE_OBS']]
plt.title(' '.join(title_string))
```

b) 15 Marks

```python
#correction for LOS
hpc_y=((np.arange(0,nx)-header['CRPIX1']-1))*header['CDELT1']/3600* np.pi / 180.
hpc_y=hpc_y[:,np.newaxis]*np.ones(nx)
B0=header['BZERO']
rsun=header['RSUN_ref']
dsun=header['DSUN_OBS']

q = dsun * np.cos(hpc_y) * np.cos(hpc_y.T)
distance = q**2 - dsun**2 + rsun**2
distance = q - np.sqrt(distance)

hcc_y=distance*np.sin(hpc_y)
hcc_x=distance*np.cos(hpc_y)*np.sin(hpc_y.T)
hcc_z=header['DSUN_OBS']-distance*np.cos(hpc_y.T)*np.cos(hpc_y)

little_r=np.sqrt(hcc_x**2+hcc_y**2+hcc_z**2)
theta=np.arcsin((hcc_y*np.cos(B0)+hcc_z*np.sin(B0))/little_r)
phi=np.arctan2(hcc_x,hcc_z)

correction=np.cos(theta)*np.cos(phi)
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(theta*180/np.pi)
ax[1].imshow(phi*180/np.pi)
```

```python
plt.imshow(correction)
plt.colorbar()
```

```python
new_data=data/correction
```

```python
plt.figure(figsize=(10,10))
plt.imshow(np.clip(new_data,-100,100),cmap='gray',origin='lower',extent=[mn_ax, mx_ax, mn_ax, mx_ax])
plt.xlabel('Solar X (arcsec)')
plt.ylabel('Solar Y (arcsec)')
```
